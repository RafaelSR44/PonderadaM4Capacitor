```python
import pandas as pd

import matplotlib.pyplot as plt

# Ler os dados do arquivo
dados = pd.read_csv('dados.csv')

# Criar o grfico
plt.figure(figsize=(10, 6))
plt.plot(dados['time'], dados['resistor'], label='Voltagem do Resistor', marker='o')
plt.plot(dados['time'], dados['capacitor'], label='Voltagem do Capacitor', marker='s')

plt.xlabel('Tempo')
plt.ylabel('Voltagem (V)')
plt.title('Variao da Voltagem do Resistor e do Capacitor')
plt.legend()
plt.grid(True)
plt.show()
```
